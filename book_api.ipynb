```python
import json

import requests

import pandas as pd
```

```python
def list_func(items):
    list_1 = []
    for sub in items:
        name = sub['name']
        list_1.append(name)
    list_1 = '; '.join(list_1)
    return list_1
```

```python
book_content = json.loads(requests.get('https://openlibrary.org/api/books?bibkeys=ISBN:9780545425117&jscmd=data&format=json').content)
book2_content = json.loads(requests.get('https://openlibrary.org/api/books?bibkeys=ISBN:9781590524367&jscmd=data&format=json').content)
book3_content = json.loads(requests.get('https://openlibrary.org/api/books?bibkeys=ISBN:9781590523667&jscmd=data&format=json').content)
```

```python
df1 = pd.DataFrame.from_dict(book_content).T
df2 = pd.DataFrame.from_dict(book2_content).T
df3 = pd.DataFrame.from_dict(book3_content).T
```

```python
df_m = pd.concat([df1, df2, df3])
df_m
```

```python
# df_m['identifiers'] = df_m['identifiers'].apply(lambda x: x.get('goodreads'))
```

```python
df_m['authors'] = df_m['authors'].apply(lambda item_list: list_func(item_list) if isinstance(item_list, list) else None)
```

```python
df_m['subjects'] = df_m['subjects'].apply(lambda item_list: list_func(item_list) if isinstance(item_list, list) else None)
```

```python
df_m['subject_people'] = df_m['subject_people'].apply(lambda item_list: list_func(item_list) if isinstance(item_list, list) else None)
```

```python
df_m['publishers'] = df_m['publishers'].apply(lambda item_list: list_func(item_list) if isinstance(item_list, list) else None)
```

```python
df_m['subject_places'] = df_m['subject_places'].apply(lambda item_list: list_func(item_list) if isinstance(item_list, list) else None)
```

```python
df_m['subject_times'] = df_m['subject_times'].apply(lambda item_list: list_func(item_list) if isinstance(item_list, list) else None)
```

```python
df_m['publish_places'] = df_m['publish_places'].apply(lambda item_list: list_func(item_list) if isinstance(item_list, list) else None)
```

```python
df_m = df_m.drop(columns=['cover', 'ebooks', 'key', 'links', 'url', 'classifications', 'by_statement', 'notes', 'pagination', 'table_of_contents'])
```
