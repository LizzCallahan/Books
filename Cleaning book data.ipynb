```python
df = pd.merge(df_10k_ratings, df_100k_books, how='left', on='isbn', indicator=True)
# Originally merged inner, but then decided to merge left instead for more data to analyze
# Might make a smaller set to analyze the overlap
```

```python
df_clean = df.drop(columns=['goodreads_book_id', # keeping book_id to match with to_read
                    'best_book_id', # keeping work_id, which is the overarching id for goodreads
                    'books_count', 
                    'isbn13_x', # keeping isbn, but droping extended isbn
                    'language_code', 
                    'image_url',# description, images, and links probably not very useful for simple analysis
                    'small_image_url', 
                    'author', # have 'authors' data intact
                    'bookformat', # going to combine across formats
                    'desc', 
                    'img', 
                    'isbn13_y', 
                    'link',  
                    'rating', # have ratings and review data from other dataframe
                    'reviews', 
                    'title_y', # have 'original_title' and 'title_x'
                    'totalratings'])
```

```python
df_clean.head()
```

```python
df_clean.describe()
```

```python
# Negative numbers in publication year don't make sense.
# Also, for modern tastes, we'll limit this to after 1960
df_clean['original_publication_year'].value_counts()
```

```python
#looking for outliers
df_clean['original_publication_year'].value_counts().plot(kind='line')
```

```python
df_clean = df_clean[df_clean['original_publication_year'] >= 1960]
```

```python
df_clean['original_publication_year'].value_counts()
```

```python
# Limiting the year lost us about 1000 books - probably could limit more, but not for now
df_clean.shape
```

```python
# It looks like about 50% of books in the set get above a 4 average rating. 
# So let's drop anything that's averaging 3.5 or below.
df_clean = df_clean[df_clean['average_rating'] >= 3.5]
```

```python
# This dropped about another 300 books. 
# Still have a pretty good number to work with.
df_clean.shape
```

```python
df_clean.isnull().sum()
```

```python
# Since some books are missing original titles, 
# but all have titles in the other column, 
# I'm dropping the original title column.
df_clean = df_clean.drop(columns='original_title')
```

```python
df_clean.head()
```

```python
# Checking to see that work ids are all unique for merging with to_read
df_clean.work_id.nunique()
```

```python
df_toread.shape
```

```python
#Groupby with book_id brings us close to the original book count on the corresponding book csv
id_grouped = df_toread.groupby(by='book_id').count().reset_index()
id_grouped
```

```python
id_grouped.rename(columns={'user_id':'to_read_count'}, inplace=True)
```

```python
id_grouped
```

```python
df_3 = pd.merge(df_clean, id_grouped, how='left', on='book_id')
df_3.head()
```

```python
# The count for how many people marked the book to_read is added.
# Not sure if it makes sense to fill null with 0 for that column yet.
df_3["to_read_count"].notnull().sum()
```

```python
df_3 = df_3.set_index('book_id')
df_3.head()
```

```python
# Check for potential duplicates with number of unique entries per column
# Looks like there are a few duplicates in the title column, 
# so we'll glance at those and see if we need to drop or combine those.
df_3.nunique()
```

```python
df_3_dup = df_3[df_3.duplicated(['title_x'], keep=False)]
```

```python
df_3.shape
```

```python
# Checking duplicates to see if I want to delete or combine things.
# Only one true duplicate - the "Salem's Lot" row. 
# Checking online, it's a significant revision, so we'll let it stand.
df_3_dup.sort_values(by="title_x")
```
